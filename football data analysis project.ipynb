```python
#import the libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load dataset

epl_df=pd.read_csv('EPL_20_21.csv')
```

```python
epl_df
```

```python
epl_df.info()
```

```python
epl_df.isna().sum()
```

```python
epl_df.describe()
```

```python
# create two new columns

epl_df['MinsPerMatch']=(epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df
```

```python
# Total goals

Total_Goals= epl_df['Goals'].sum()
Total_Goals
```

```python
# Penalty Goals

Total_PenaltyGoals=epl_df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
# penalty Attemts

Total_Penalty_attempts=epl_df['Penalty_Attempted'].sum()
Total_Penalty_attempts
```

```python
# pie chart for penalties missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored=epl_df['Penalty_Attempted'].sum()-Total_PenaltyGoals
data=[pl_not_scored,Total_PenaltyGoals]
labels=['Penalties Missed','Penalties Scored']
plt.pie(data,labels=labels,autopct='%.0f%%')
plt.show()
```

```python
# unique positions

epl_df['Position'].unique()
```

```python
# total FW players

epl_df[epl_df['Position']=='FW']
```

```python
# players from different nations 

np.size((epl_df['Nationality'].unique()))
```

```python
# players from different nations 

nationality=epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12, 6),color=sns.color_palette("magma"))
plt.show()
```

```python
# clubs with maximum players in their squard

epl_df['Club'].value_counts().nlargest(30).plot(kind='bar',color=sns.color_palette("viridis"))
```

```python
# clubs with least players in their squard

epl_df['Club'].value_counts().nsmallest(30).plot(kind='bar',color=sns.color_palette("viridis"))
```

```python
# Players based on age group

Under20=epl_df[epl_df['Age']<=20]
Under20_25=epl_df[(epl_df['Age']>20) & (epl_df['Age']<=20)]
Under25_30=epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
Above30=epl_df[epl_df['Age']>30]
```

```python
plt.figure(figsize=(13,6))
x=np.array([Under20['Name'].count(),Under20_25['Name'].count(),Under25_30['Name'].count(),Above30['Name'].count()])
mylabels=["<=20",">=20 & <=25",">=25 & <=30",">30"]
plt.title('total Players With Age',fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Under20 players in each club 

players_under_20 = epl_df[epl_df['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("cubehelix"))
```

```python
# Under 20 players in Manchester United 

players_under_20[players_under_20["Club"]=='Manchester United']
```

```python
# Under 20 players in Chelsea

players_under_20[players_under_20["Club"]=='Chelsea']
```

```python
# Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=epl_df)
plt.xticks(rotation=90)
```

```python
num_player=epl_df.groupby('Club').size()
data=(epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
#Total assists from each club

Assists_by_clubs=pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by="Assists"),palette="Set2")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=30)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assits',fontsize=30)
plt.show()
```

```python
# top 10 Assists

top_10_assists=epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assists
```

```python
Goals_by_clubs=pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by="Goals"),palette="rocket")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=30)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=20)
```

```python
#most goals by players

top_goal_scorer=epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_goal_scorer
```

```python
# goals per match

top_10_goal_scorer_per_match=epl_df[['Name','Club','GoalsPerMatch','Matches','Goals']].nlargest(n=10,columns='GoalsPerMatch')
top_10_goal_scorer_per_match
```

```python
# pie Chart -Goals with most assists and without assists

plt.figure(figsize=(14,6))
assists=epl_df['Assists'].sum()
data=[Total_Goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```
